```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([3,4,2,4,5])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
x_mean=x.mean()
y_mean=y.mean()
```

```python
numer=0
deno=0
for i in range(len(x)):
   numer+=((x[i]-x_mean)*(y[i]-y_mean)) 
   deno+=(x[i]-x_mean)**2
m=numer/deno
m
```

```python
# y=mx+c
c=y_mean-(m*x_mean)
c
```

```python
print(f"Line Eqn: y={m}*x+{c}")
```

```python
ypred=x*m+c
plt.scatter(x,y)
plt.plot(x,ypred,c='r')
```

```python
ypred
```

```python
numer=0
deno=0
for i in range(len(x)):
    numer+=(ypred[i]-y_mean)**2
    deno+=(y[i]-y_mean)**2
numer/deno
```

```python
class LinearRegression:
    def __init__(self):
        self.x=0
        self.y=0
        self.coef_=0
        self.intercept_=0
        self.x_mean=0
        self.y_mean=0
    def fit(self,x,y):
        self.x=x
        self.y=y
        self.x_mean=self.x.mean()
        self.y_mean=self.y.mean()
        numer=0
        deno=0
        for i in range(len(x)):
           numer+=((x[i]-self.x_mean)*(y[i]-self.y_mean)) 
           deno+=(x[i]-self.x_mean)**2
        self.coef_=numer/deno
        self.intercept_=y_mean-(m*self.x_mean)
    def predict(self,X):
        return X*self.coef_+self.intercept_
        
    def diagram(self):
        ypred=self.predict(self.x)
        plt.scatter(self.x,self.y)
        plt.plot(self.x,ypred,c='r')
    def error(self):
        ypred=self.predict(self.x)
        numer=0
        deno=0
        for i in range(len(self.x)):
            numer+=(ypred[i]-self.y_mean)**2
            deno+=(self.y[i]-self.y_mean)**2
        return numer/deno
    def accuracy(self):
        return 1-self.error()
        
    def __repr__(self):
        return f"model equation: y={self.coef_}*x+{self.intercept_}"
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```
