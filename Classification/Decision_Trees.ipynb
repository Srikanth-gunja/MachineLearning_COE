
# Decision Tree: Quick Guide

## 1. Introduction
A **Decision Tree** is a supervised learning algorithm used for classification and regression. It splits data into subsets based on feature values.

## 2. Steps to Build a Decision Tree
1. **Select the Best Feature**: Choose the feature with the highest Information Gain (IG) or lowest impurity.
2. **Split the Data**: Divide the dataset based on the selected feature.
3. **Repeat Recursively**: Continue splitting until stopping criteria (e.g., pure nodes, max depth) is met.
4. **Prune the Tree (Optional)**: Reduce complexity to prevent overfitting.

## 3. Key Concepts
- **Entropy**: Measures disorder in data.
  $$
  Entropy = - \sum p_i \log_2 p_i
  $$
- **Gini Index**: Measures impurity.
  $$
  Gini = 1 - \sum p_i^2
  $$
- **Information Gain**: Reduction in entropy after a split.
  $$
  IG = Entropy(parent) - \sum \left(\frac{|subset|}{|parent|} \times Entropy(subset) \right)
  $$
## 4. Advantages & Disadvantages
✔ Easy to understand and interpret.  
✔ Works with numerical and categorical data.  
❌ Prone to overfitting.  
❌ Sensitive to small data variations.

## 5. Applications
- **Medical Diagnosis**
- **Spam Detection**
- **Credit Risk Assessment**
- **Recommendation Systems**

## 6. Conclusion
Decision Trees are simple yet powerful models. To improve performance, consider **pruning** or using **ensemble methods** like Random Forest.
